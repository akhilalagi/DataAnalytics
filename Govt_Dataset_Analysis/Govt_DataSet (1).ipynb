#                                 Indra Gandhi Widow Pension Scheme FY 2023-2024

The **Indira Gandhi Widow Pension Scheme (2023-2024) ensures structured tracking of welfare benefits across regions and demographics. LGD codes uniquely identify states and districts, while scheme codes categorize programs. This enhances transparency and accountability, ensuring funds reach the right beneficiaries. Aadhaar and mobile numbers aid in DBT validation, preventing fraud. Demographic analysis (SC, ST, OBC, General) ensures fair distribution.Time-based tracking (financial year, month, last update) allows trend analysis and irregularity detection. Covering multiple districts in Jammu & Kashmir the dataset provides a comprehensive and structured approachto welfare monitoring.

```python
aadhar_data = [df['total_aadhar'].sum(), df['total_beneficiaries'].sum() - df['total_aadhar'].sum()]
labels = ["With Aadhaar", "Without Aadhaar"]

plt.figure(figsize=(8, 8))
plt.pie(aadhar_data, labels=labels, autopct='%1.1f%%', colors=['green', 'yellow'])
plt.title("Aadhaar vs Non-Aadhaar Beneficiaries")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['total_aadhar'], vert=False)
plt.title("Variation of Aadhaar Registrations Across Districts")
plt.xlabel("Total Aadhaar Registrations")
plt.grid(True)
plt.show()
```

```python
top_districts = df.groupby('district_name')['total_beneficiaries'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
top_districts.plot(kind='bar', color='coral')
plt.title("Top 10 Districts with Most Beneficiaries")
plt.xlabel("Districts")
plt.ylabel("Total Beneficiaries")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
bottom_states = df.groupby('state_name')['total_beneficiaries'].sum().nsmallest(10)

plt.figure(figsize=(12, 6))
bottom_states.plot(kind='barh', color='brown')
plt.title("States with Lowest Beneficiaries")
plt.xlabel("Total Beneficiaries")
plt.ylabel("State")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
mobile_data = [df['total_mobileno'].sum(), df['total_beneficiaries'].sum() - df['total_mobileno'].sum()]
labels = ["With Mobile Number", "Without Mobile Number"]

plt.figure(figsize=(8, 8))
plt.pie(mobile_data, labels=labels, autopct='%1.1f%%', colors=['green', 'gray'])
plt.title("Mobile Number Registration Among Beneficiaries")
plt.show()
```

```python
state_name = "MAHARASHTRA"  # Change to the desired state
state_data = df[df['state_name'] == state_name][['total_aadhar', 'total_beneficiaries']].sum()
state_data['non_aadhar'] = state_data['total_beneficiaries'] - state_data['total_aadhar']

plt.figure(figsize=(8, 8))
plt.pie([state_data['total_aadhar'], state_data['non_aadhar']], labels=["With Aadhaar", "Without Aadhaar"], autopct='%1.1f%%', colors=['blue', 'gray'])
plt.title(f"Aadhaar Penetration in {state_name}")
plt.show()
```

```python
top_obc_states = df.groupby('state_name')['total_obc'].sum().nlargest(5)

plt.figure(figsize=(10, 5))
top_obc_states.plot(kind='bar', color='brown')
plt.title("Top 5 States with Highest OBC Beneficiaries")
plt.xlabel("State")
plt.ylabel("Total OBC Beneficiaries")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
