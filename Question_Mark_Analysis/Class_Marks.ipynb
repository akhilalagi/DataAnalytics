Dataset Description
This dataset contains student marks for different questions in an assessment. It includes 86 records with 12 columns, where each column represents either a specific question's marks or the total score. Some columns have missing values, indicating that not all students attempted every question.

Dataset Overview
Total Marks: Represents the overall score of each student.

Q1aM4, Q1bM6, Q2aM6, etc.: Each column represents marks obtained for a specific question, with the maximum possible marks indicated in the column name (e.g., "Q1aM4" means question 1a with a max score of 4).

Missing Values: Some questions have a significant number of missing values, likely due to optional or unattempted questions.

Data Type: Numeric (integer/float).
